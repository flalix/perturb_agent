### Paper

Tahoe-x1: Scaling Perturbation-Trained Single-Cell Foundation Models to 3 Billion Parameters

https://www.biorxiv.org/content/10.1101/2025.10.23.683759v1.full

New blog post with our friends at NVIDIA Healthcare: Drugs do not affect just a single target. Their effect propagate through the gene regulatory network (GRN), producing cascades of downstream transcriptional changes. 

But mapping these effects across a 100M-cell drug-perturbation is not tractable with standard computers. We teamed up with the NVIDIA team to solve that problem and we discussed that in our new blog post. 

We introduce Tx1-CD, a causal decoder trained on Tahoe-x1 embeddings derived from our Tahoe-100M dataset. TX1-CD decomposes each drug response into a proximal signature and a propagated cascade through a learned directed GRN, producing a 14K gene network designed to explain how perturbation effects move through regulation, not just which genes co-vary.

What enabled this is building it GPU-native. We run this pipeline on NVIDIA H200 with CUDA-X Data Science and rapids-singlecell, with the next phase pushing further into multi-node training on next-generation GPUs, including B200, using distributed sharding approaches such as nvFSDP.

On the same masked edge set, TX1-CD learns structures that raw coexpression does not. Its edge rankings align substantially better with external references. 

The payoff is mechanistic usability: TX1-CD turns signatures into explicit cascades, it highlights candidate regulators and bottlenecks that mediate drug effects, and provides a way to generate testable hypotheses about convergence between perturbations and about regulatory edges that are not captured in existing references.
