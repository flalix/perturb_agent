```python
set_page_config(page_title="Reactome Explorer", layout="wide")

st.title("🧬 Reactome Pathway Explorer")

root_data='../data/reactome/'
verbose = True

rea = Reactome(root_data=root_data)

def open_tables(verbose:bool=True):
 
    _ = rea.open_reactome(verbose=verbose)
    _ = rea.open_reactome_abstract(verbose=verbose)
    _ = rea.open_reactome_gmt(verbose=verbose)

    return

def tool1(pathway: str) -> Tuple[str, List]:

    df_gmt = rea.df_gmt

    row = df_gmt[df_gmt.pathway == pathway].iloc[0]

    genes = row.genes
    if isinstance(genes, str):
        genes = eval(genes)

    return row.pathway_id, genes

open_tables(verbose=verbose)

REACTOME_PATHWAYS = rea.pathway_list

# --- Dropdown selection ---
selected_pathway = st.selectbox(
    "Select a Reactome pathway:",
    REACTOME_PATHWAYS
)

# --- Button to trigger computation ---
if st.button("Run analysis"):
    with st.spinner(f"Running analysis for: {selected_pathway}..."):

        # Call your backend tool
        result = tool1(pathway=selected_pathway)

    st.success("Pathway found ...")

    # --- Display results ---
    st.subheader("Pathway")

    # Flexible rendering depending on output
    if isinstance(result, dict):
        st.json(result)
    elif hasattr(result, "head"):  # pandas DataFrame
        st.dataframe(result)
    else:
        st.write(result)

!streamlit run app.py --server.port 8502        
```
