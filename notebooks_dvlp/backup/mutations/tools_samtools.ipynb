### Tabix

Tabix is not part of the bcftools executable itself, but it is a closely related tool bundled within the same HTSlib library package. While bcftools manages variant manipulation, tabix is used to index tab-delimited genomic files (like VCF or BED) for fast random access, often producing a .tbi file. 

Tabix was written by Heng Li. The BGZF library was originally implemented by Bob Handsaker and modified by Heng Li for remote file access and in-memory caching.





### Run tabix (an example)

- change diretory <your dir>

```Bash
sudo apt install tabix

cd ../VCF/parkinson/ukb-b-16943-sibling  
ls -ls *.vcf.gz  
tabix -p vcf ukb-b-16943.vcf.gz  
tabix -p vcf ukb-b-16943_raw.vcf.gz
```



### Run VEP on your VCF

```Bash
vep \
  -i UKB-b-16943.vcf.gz \
  -o annotated.txt \
  --cache \
  --assembly GRCh38 \
  --vcf
```

### Better command (for GWAS)

```Bash
vep \
  -i UKB-b-16943.vcf.gz \
  -o annotated.vcf \
  --cache \
  --assembly GRCh38 \
  --vcf \
  --symbol \
  --nearest symbol \
  --distance 10000 \
  --everything
```
