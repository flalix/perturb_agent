```python
# import libraries
import os
from IPython.display import Markdown, display
from lightrag import LightRAG, QueryParam
from lightrag.llm.openai import gpt_4o_mini_complete, gpt_4o_complete, openai_embed
from lightrag.kg.shared_storage import initialize_pipeline_status
import PyPDF2
from google.colab import userdata
```

# Create a directory

```python
# Set the name for the working directory
WORKING_DIR = "./working_dir"

if not os.path.exists(WORKING_DIR):
    os.makedirs(WORKING_DIR)
```

```python
# Create the LightRAG configurations
rag = LightRAG(
    working_dir=WORKING_DIR,
    embedding_func = openai_embed,
    llm_model_func= gpt_4o_mini_complete
)
```

```python
# If you are using Colab / Jupyter
import nest_asyncio
nest_asyncio.apply()
await rag.initialize_storages()  # Initialize storage backends
await initialize_pipeline_status()  # Initialize processing pipeline
```

# Data Preparation


*   Open the file
*   transform the PDF into text - Transcription
*  Perform the RAG



```python
# Open the PDF and extract the text
pdf_text = []
with open("Greek-gods-story.pdf", "rb") as file:
  reader = PyPDF2.PdfReader(file)
  for page in reader.pages:
    pdf_text.append(page.extract_text())
pdf_text
```

```python
# Save the data
with open("greek_gods.txt", "w") as file:
  for line in pdf_text:
    file.write(line)
```

```python
# Build the RAG
with open("greek_gods.txt", "r") as file:
  rag.insert(file.read())
```

# RAG

* Naive
*  Local
*  Global



```python
from lightrag.utils import setup_logger

setup_logger("lightrag", level = "WARNING")
```

```python
# Perform the RAG
query = "Who is Aphrodite"
def perform_rag(query, mode):
  return Markdown(rag.query(
    query = query,
    param = QueryParam(
        mode = mode,
        enable_rerank = False
    )
))
```

```python
# Performing with Local
perform_rag(query, "local")
```

```python
# Perform RAG with global
perform_rag(query, "global")
```

```python
# Perform RAG with hybrid
perform_rag(query, "hybrid")
```

# Graph Visualization

```python
# import libraries
import networkx as nx
from pyvis.network import Network
import matplotlib.pyplot as plt
import random
from IPython.display import HTML
```

```python
# Load entities
G = nx.read_graphml(f"{WORKING_DIR}/graph_chunk_entity_relation.graphml")

# Create the network
net = Network(
    notebook = True,
    cdn_resources = "remote")
net.from_nx(G)

# add colors and titles to nodes
for node in net.nodes:
  node["color"] = "#{:06x}".format(random.randint(0, 0xFFFFFF))
  if "description" in node:
    node["title"] = node["description"]

# Add titles to edges
for edge in net.edges:
  if "description" in edge:
    edge["title"] = edge["description"]

# Save the file
net.show("graph.html")
net.save_graph("graph.html")

# Display
HTML(filename = "graph.html")
```

```python
# Visualize the network with a target node
# Load entities
G = nx.read_graphml(f"{WORKING_DIR}/graph_chunk_entity_relation.graphml")

# Target node
target_node = "Paris"

# Get the neighbors
subgraph_nodes = list(G.neighbors(target_node)) + [target_node]
subgraph = G.subgraph(subgraph_nodes)

# Create the network
net = Network(
    notebook = True,
    cdn_resources = "remote")
net.from_nx(subgraph)

# add colors and titles to nodes
for node in net.nodes:
  node["color"] = "#{:06x}".format(random.randint(0, 0xFFFFFF))
  if "description" in node:
    node["title"] = node["description"]

# Add titles to edges
for edge in net.edges:
  if "description" in edge:
    edge["title"] = edge["description"]

# Save the file
net.show("graph.html")
net.save_graph("graph.html")

# Display
HTML(filename = "graph.html")
```
