# Relatório Científico — Análise Transcriptômica da Inflamação Cardíaca na Doença de Chagas
## Composição vs. Volume: Teste Empírico da Hipótese em Dados Públicos de Biópsia Miocárdica

**Preparado por:** Prometheus Lab (CFIB)
**Data:** 16 de junho de 2026
**Versão:** 2.0 — resultados da expedição computacional exp-chagas-composition-endpoint
**Destinatária:** Prof.ª Ester Sabino — Instituto de Medicina Tropical, USP
**Referência ao documento anterior:** Relatório CFIB-Lab-Sabino-v1 (junho 2026)

---

## 1. Sumário Executivo

Este relatório apresenta os resultados de uma expedição de bioinformática que testou empiricamente, em dados transcriptômicos públicos de biópsia miocárdica humana, a hipótese central do relatório anterior: que a **composição** da inflamação cardíaca na Doença de Chagas — especificamente a razão entre o braço citotóxico deletério (linfócitos CD8⁺ com perforina e granzima) e o braço protetor (células produtoras de IFN-γ) — prediz severidade melhor do que o simples **volume** total de infiltrado.

A análise utilizou dois conjuntos de dados públicos independentes (GSE191081, RNA-sequenciamento; GSE84796, microarray Agilent) totalizando 39 amostras de miocárdio ventricular humano. Nenhum novo paciente foi recrutado.

**Resultado principal:** A desproporção entre o braço citotóxico e o protetor é real, Chagas-específica e replicada em duas plataformas. O braço deletério (GZMB/PRF1) cresce 5,7 vezes mais em CCC grave em relação ao controle, enquanto o protetor (IFNG/CXCL9) cresce apenas 1,75 vezes. Contudo, com apenas 8 casos de CCC grave, não foi possível demonstrar que a **razão** de composição é um preditor estatisticamente independente do **volume** total de infiltrado — os dois eixos são altamente colineares (r = 0,926) e matematicamente inseparáveis com o N disponível.

Dois achados adicionais relevantes emergem: (1) um infiltrado plasmocitário tecidual massivo e Chagas-específico, com expressão de IgG 565 vezes maior em CCC que em controles, confirmado como tecidual por controle negativo; e (2) dois subtipos moleculares distintos entre os 8 casos de CCC grave, cuja nomenclatura clínica aguarda correlação com dados de FEVE/forma clínica ainda não disponíveis publicamente.

## 2. Contexto e Motivação

O relatório anterior propôs que a pergunta terapêutica central para Chagas cardíaca não é "quanto de inflamação" mas "de que tipo de inflamação" — especificamente, que o excesso relativo de células citotóxicas (perforina⁺/granzima⁺) sobre células protetoras (IFN-γ⁺) define prognóstico e resposta a intervenção.

Esta hipótese tem suporte conceitual forte na literatura sobre cardiomiopatia chagásica crónica (CCC). No entanto, nunca foi testada diretamente em dados transcriptômicos de biópsia miocárdica humana com análise formal de composição versus volume.

Antes de desenhar um novo estudo clínico — que demandaria recrutamento, biópsia invasiva e anos de execução — decidimos testar a hipótese nos dados já disponíveis. Dois datasets de RNA-seq e microarray de biópsia miocárdica humana em CCC estão publicamente acessíveis no Gene Expression Omnibus (GEO/NCBI). A análise foi concluída em aproximadamente 24 horas, sem custo de amostragem.

**GSE191081** é o dataset principal. Contém três grupos: controles saudáveis (CTRL), cardiomiopatia dilatada não-chagásica (DCM) — insuficiência cardíaca terminal de outra causa — e CCC grave (sevCCC). **GSE84796** é o dataset de validação cruzada. Foi gerado em plataforma tecnologicamente independente (microarray em vez de sequenciamento), em coorte diferente de pacientes. Serve para confirmar que os achados do GSE191081 não são artefato de uma única tecnologia ou coorte.

**Nota de transparência metodológica:** Ambos os datasets foram originalmente processados com exclusão de genes de imunoglobulina e receptor de células T (Ig/TCR), tratados como ruído de contaminação por células sanguíneas circulantes. Para esta análise, **re-incluímos deliberadamente estes genes**, pois a hipótese do R13 (infiltrado plasmocitário produtor de IgG anti-ADRB1) exige exatamente esta categoria. A re-inclusão foi aplicada igualmente a todos os grupos, sem viés direcional. Os achados de IGHG1 foram validados contra o grupo DCM (ver Seção 6.1, Achado 2) para descartar contaminação sanguínea.

## 5. Métodos de Análise

### 5.1 Normalização dos dados

**RNA-sequenciamento (GSE191081):** os dados brutos de contagem de genes foram normalizados para CPM (contagens por milhão de leituras), depois transformados para log(CPM+1). CPM é uma forma de corrigir diferenças no total de leituras entre amostras — permite comparar expressão entre pacientes como se todos tivessem o mesmo "tamanho" de biblioteca de sequenciamento.

**Microarray Agilent (GSE84796):** plataforma one-color (canal único, marcação Cy3). Os valores de intensidade já vinham normalizados pelos autores por "percentile-shift" — uma técnica que ajusta cada amostra para ter a mesma mediana global, permitindo comparação entre chips. Os valores são em escala log2. Para calcular diferença entre grupos, usamos: FC = média(CCC) − média(CTRL) em escala log2, que equivale a um log-ratio.

### 5.2 Scores compostos

Para cada amostra calculamos três scores:

- **D_score (braço Deletério):** média de log(CPM+1) dos genes citotóxicos principais — GZMB (granzima B), PRF1 (perforina), GZMH e GZMK (granzimas H e K)
- **P_score (braço Protetor):** média de log(CPM+1) dos genes protetores — IFNG (interferon-gama), GBP2, CXCL9, STAT1, IRF1
- **V_score (Volume de infiltrado):** média de log(CPM+1) dos marcadores pan-T — CD8A, CD8B, CD3D, CD3E

A **razão de composição** foi calculada como D_score / P_score (ou D_score − P_score em escala log, que é equivalente).

### 5.3 Teste de colinaridade e informação independente

Para verificar se a razão D/P adiciona informação preditiva além do volume, fizemos dois testes:

1. **Resíduo de volume:** calculamos o resíduo da razão D/P depois de remover o efeito linear do V_score (regressão linear: razão ~ volume). Se a composição fosse informativa independente do volume, estes resíduos deveriam separar grupos CCC de CTRL. Não separaram (Mann-Whitney p=0,195).

2. **Critério de informação de Akaike (AIC):** comparamos dois modelos logísticos (CCC vs outros): (a) só com V_score; (b) com V_score + razão D/P. Se a razão adicionasse informação, o AIC do modelo (b) deveria ser menor. O AIC piorou em 1,94 — a razão não ajudou.

3. **Teste de permutação:** para verificar se a diferença entre d(razão)=5,35 e d(volume)=3,31 é estatisticamente real, embaralhamos aleatoriamente os rótulos de grupo 10.000 vezes. A diferença observada ocorreu em apenas 1 de 10.000 permutações (p=0,0001) — a assimetria de magnitude é real. Contudo, real não implica dissociável.

### 5.4 Controle negativo DCM (Decoy)

O grupo DCM (cardiomiopatia dilatada não-chagásica) serve como controle negativo "nativo". Estes pacientes têm insuficiência cardíaca terminal severa — se o padrão observado em CCC fosse apenas reflexo de coração insuficiente, DCM deveria apresentar o mesmo padrão. Adicionalmente, DCM tem mais células sanguíneas retidas no tecido (maior contaminação por sangue), o que é útil para testar se sinais de plasma/soro aparecem nos dados.

### 5.5 Validação cruzada entre plataformas

A validação entre RNA-seq e microarray foi **estritamente direcional**: só comparamos o sinal (UP ou DOWN) e a ordenação relativa (braço D > braço P, Track A > Tier 2). Não combinamos p-values nem FCs entre plataformas, pois as escalas absolutas são incomparáveis.

## 6. Resultados

### 6.1 O que foi Confirmado

#### Achado 1: Assimetria de magnitude do braço citotóxico — confirmado em 2 plataformas

No GSE191081 (RNA-seq), os 8 casos de CCC grave mostram:
- D_score: +5,7× em relação a CTRL (variação enorme)
- P_score: +1,75× em relação a CTRL (variação modesta)
- O braço citotóxico cresce mais que 3 vezes mais rápido que o protetor

No GSE84796 (Agilent), os 10 casos de CCC:
- FC médio genes deletérios (GZMB, PRF1, GZMH, GZMK) = **4,15 log2**
- FC médio genes protetores (IFNG, GBP2, CXCL9, STAT1, IRF1) = **2,50 log2**
- Shift de composição D−P: **+1,65 log2** em CCC vs −1,46 em CTL

**100% de concordância direcional** (9/9 genes testados UP em CCC em ambas as plataformas).

A permutação confirma que a diferença entre d(D_score) e d(P_score) não é acaso (p=0,0001 em 10.000 permutações).

#### Achado 2 (mais limpo da expedição): Infiltrado plasmocitário tecidual massivo e Chagas-específico

IGHG1 (imunoglobulina G tipo 1, produzida por plasmócitos maduros) em RNA-seq:
- CTRL: ~4 CPM
- DCM: ~7,9 CPM
- sevCCC: **2.397 CPM** — 565 vezes mais que CTRL

**Prova de que é tecidual e não contaminação de sangue:**

O grupo DCM é o controle decisivo. DCM tem 35 vezes mais hemoglobina nos dados (HBB = 557 CPM) que controles — evidência de maior contaminação por sangue no tecido DCM. Se o IGHG1 alto em CCC fosse simplesmente sangue retido no tecido, DCM deveria ter IGHG1 também alto (tem mais sangue). Mas DCM tem IGHG1 = 7,9 CPM — quase nulo. CCC tem IGHG1 = 2.397 CPM com HBB mais baixo que DCM.

Adicionalmente: IGHG1 correlaciona com SDC1/CD138 (marcador de superfície de plasmócito, r=0,903) e com PTPRC/CD45 (leucócito, r=0,732), e *anti*correlaciona com hemoglobina HBB (r=−0,243). O perfil é de **células plasmocitárias infiltradas no tecido**, não de sangue.

Confirmação no Agilent: SDC1 UP (+1,87 log2) e IGJ (cadeia J de imunoglobulina, marcador de plasmócito secretor) UP (+6,18 log2) — comparável em magnitude a GZMH (+6,12) e CXCL9 (+6,00).

#### Achado 3: Assimetria de composição é Chagas-específica, não end-stage genérico

Shift de composição (D_score − P_score) por grupo:
- CTRL: −1,46
- **DCM**: −1,44 (praticamente igual ao CTRL — shift ≈ +0,022)
- **sevCCC**: −0,19 (shift ≈ +0,197 em relação ao CTRL — **9× maior que DCM**)

Um coração em falência terminal não exibe este padrão. A desproporção citotóxico>protetor é biologia de Chagas.

#### Achado 4: Hierarquia Track A >> Tier 2 — alvos farmacológicos secundários não são o driver

Os alvos farmacológicos de segunda linha propostos no relatório anterior (estatinas via HMGCR/ROCK, pentoxifilina via PDE4/cAMP) respondem por menos de 3% da variância explicada pelo braço citotóxico:

No RNA-seq: variância de Tier 2a (ROCK/NF-κB) = 1,0% de R²(Track A); Tier 2b (PDE4/cAMP) = 2,4%.

No Agilent: GZMB FC=2,86 e CXCL9 FC=6,00 vs ROCK1 FC=0,24, PDE4B FC=0,29. Vários genes Tier 2 estão flat ou levemente negativos no array.

**Conclusão:** as vias de ROCK/PDE4/NF-κB não são o driver primário da cardiomiopatia — são fenômenos secundários ou consequências do dano citotóxico. O foco primário de intervenção permanece no braço citotóxico.

### 6.2 O que foi Refutado

#### Refutação 1: A hipótese forte "composição prediz melhor que volume" — não sustentada com N=8

Esta é a correção mais importante em relação ao relatório anterior.

A hipótese de que a razão D/P seria um preditor *independente e superior* ao volume total de infiltrado não pode ser sustentada com os dados disponíveis. Os dois eixos são geometricamente quase idênticos:

- Correlação entre V_score e razão D/P: **r = 0,926** (muito alta)
- Após remover o efeito do volume (resíduo): razão não separa CCC de CTRL (Mann-Whitney p=0,195; Cohen's d cai de 5,35 para 0,76)
- Modelo logístico com razão + volume não melhora sobre modelo só com volume (ΔIC = −1,94 — pior)

**Interpretação correta:** com N=8 e dados bulk de RNA de tecido inteiro, "quantas células T estão lá" e "qual proporção é citotóxica" são medidas essencialmente equivalentes. Para dissociá-las, seria necessário N > 30 casos ou análise de célula única (single-cell RNA-seq), que resolve cada célula individualmente.

**O que se mantém:** a *descrição biológica* da assimetria é real e Chagas-específica. A hipótese de que esta assimetria tem valor preditivo *independente* do volume total aguarda coorte maior.

#### Refutação 2: ARG1:NOS2 (Filtro P1 do relatório anterior) — não detectável em RNA-seq humano

| Gene | Amostras com CPM > 1 (de 22) |
|------|-------------------------------|
| NOS2 (óxido nítrico sintase 2) | 0/22 |
| ARG1 (arginase 1) | 2/22 |
| IL12B | 0/22 |

Estes genes são marcadores robustos de polarização de macrófagos M1/M2 em **camundongos**. Em humanos, a biologia é mais complexa: ARG1 e NOS2 não se comportam como marcadores equivalentes *in vivo* em tecido cardíaco humano.

**Implicação para o relatório anterior:** o "Filtro P1" (ARG1:NOS2 como indicador de atividade macrofágica protetora vs deletéria) não pode ser avaliado em dados bulk de RNA-seq humano com os genes propostos. Marcadores alternativos funcionais em humano seriam CD163/MSR1 (M2-like, 22/22 amostras detectáveis) e CD68 (pan-macrófago).

### 6.3 Achados Exploratórios (exigem validação futura)

#### Exploratório 1: Dois endótipos moleculares em CCC grave

Análise de agrupamento não-supervisionado dos 8 casos de CCC grave revela dois subgrupos:

- **Silhouette score observado:** 0,40 (quanto mais próximo de 1, mais distintos os grupos)
- **Teste de permutação** (10.000 embaralhamentos aleatórios): p = 0,0047
- **Estabilidade bootstrap:** 90,5% (os grupos são estáveis ao reamostrar)

Os dois subgrupos têm perfis distintos:
- **Subgrupo A:** alta atividade citotóxica (GZMB muito alto), atividade plasmocitária moderada
- **Subgrupo B:** altíssima atividade plasmocitária (IGHG1 dominante), citotóxico moderado

**Limitação crítica:** o GSE191081 não disponibiliza FEVE (fração de ejeção ventricular esquerda) nem forma clínica (arritmogênica vs forma dilatada) por paciente nos metadados públicos. Sem estes dados, não é possível nomear os endótipos clinicamente — não sabemos se Subgrupo A é mais arritmogênico e Subgrupo B mais dilatado, por exemplo.

**Ação necessária:** os dados clínicos dos 8 pacientes de CCC do GSE191081 podem estar disponíveis com os pesquisadores responsáveis pelo dataset. Contato sugerido com Jean-Michel Chevillard (INSERM-U1090) ou Edecio Cunha-Neto (InCor/USP) pode desblocar esta análise.

#### Exploratório 2: Co-dominância humoral — hipótese nova emergente

Os dados sugerem que, no end-stage da CCC, **dois braços imunes co-dominam simultaneamente**: o braço citotóxico (GZMB/PRF1/GZMH) e um braço humoral-autoimune (plasmócitos secretores de IgG).

No Agilent, IGJ (cadeia J, produzida por plasmócitos ativos secretando imunoglobulina) tem FC = +6,18 log2 — comparável a GZMH (+6,12) e CXCL9 (+6,00), os maiores sinais citotóxicos.

**Hipótese:** o relatório anterior organizava citotóxico e autoimune numa hierarquia (citotóxico >> autoimune). Os dados sugerem que eles podem ser **complementares e co-dominantes** no estágio terminal — não hierárquicos.

**Caveat importante:** a magnitude bruta de IGHG1 é inflada pela alta taxa de transcrição de IgG por plasmócito individual (uma célula plasmocitária produz enormes quantidades de mRNA de anticorpo). Sem dados de single-cell, não sabemos se há 10 plasmócitos muito ativos ou 10.000. O número de células não é determinável com RNA-seq de tecido inteiro.

#### Exploratório 3: Braço autoimune R13 (anti-β1)

- ADRB1 (receptor β1 adrenérgico, alvo do R13): DOWN em ambas as plataformas (RNA-seq: FC=−0,24, não-significativo; Agilent: FC=−1,14, mais nítido)
- Correlação parcial plasmócito-ADRB1 (excluindo efeito de infiltrado total): r=0,375 — passa o limiar pré-definido (≥0,30), mas p=0,086 (não-significativo com N=8)
- **Interpretação:** o componente plasmocitário do R13 é robusto; ADRB1 isolado como marcador de severidade é sinal fraco em RNA de bulk tecidual. O mecanismo R13 (anticorpos anti-β1 causando disfunção adrenérgica) permanece biologicamente plausível mas exige validação funcional.

## 7. Implicações para Estratégia Terapêutica

Com base nos novos achados, as recomendações do relatório anterior são atualizadas:

### 7.1 O que se mantém

**Alvo primário no braço citotóxico deletério (GZMB/PRF1):** confirmado em dois datasets, tecnologias independentes, com especificidade Chagas (DCM não exibe o padrão). A intervenção que reduz seletivamente a atividade citotóxica de CD8⁺ perforina⁺ permanece o alvo mais fundado empiricamente.

**R13 — tratamento do componente autoimune:** o sinal plasmocitário tecidual massivo e Chagas-específico reforça que há atividade de anticorpos no miocárdio CCC. A hipótese de que anticorpos anti-ADRB1 contribuem para disfunção cardíaca merece investigação funcional direta.

### 7.2 O que precisa ser corrigido

**Filtro P1 (ARG1:NOS2) — não aplicável em RNA-seq humano:** os marcadores de polarização macrofágica M1/M2 do relatório anterior são baseados em biologia murina. Em humano, estes genes não são detectáveis em tecido cardíaco por RNA-seq bulk. Não há evidência transcriptômica de que o eixo macrofágico ARG1/NOS2 é um contribuinte relevante detectável nos dados públicos disponíveis.

**Composição vs. volume — reformulação necessária:** a frase "a composição prediz melhor que o volume" deve ser reformulada para: "a composição da inflamação é assimétrica (mais citotóxica que protetora) em CCC grave — esta assimetria é Chagas-específica e replicável, mas com os dados disponíveis (N=8 bulk), volume e composição são informativamente equivalentes." A hipótese de superioridade preditiva da composição permanece válida como hipótese a ser testada em coorte maior.

### 7.3 O que os novos achados adicionam

**Dois endótipos potenciais:** se os subgrupos moleculares detectados correspondem a formas clínicas distintas (hipótese plausível: Subgrupo A = forma arritmogênica; Subgrupo B = forma dilatada com componente autoimune proeminente), as estratégias terapêuticas deveriam ser personalizadas. Pacientes com dominância humoral (Subgrupo B) podem necessitar de abordagem que inclua imunomodulação de plasmócitos além do braço citotóxico.

**Co-dominância citotóxico + humoral:** o tratamento focado exclusivamente no braço citotóxico pode ser insuficiente se o braço plasmocitário-autoimune é igualmente destrutivo em subgrupo de pacientes. Esta é uma hipótese que justifica experimento dedicado.
