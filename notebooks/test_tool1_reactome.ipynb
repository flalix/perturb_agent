```python
# import requests
# import json
import os, sys
import pandas as pd

from pathlib import Path
from typing import List, Tuple #Optional, Iterable, Set, Any,

ROOT = Path().resolve().parent.parent
SRC = os.path.join(ROOT, "src")

if str(SRC) not in sys.path:
    sys.path.append(str(SRC))

print("ROOT:", ROOT)
print("SRC added:", SRC)

from libs.Basic import *
from libs.reactome_lib import *

root_data='../data/reactome/'
```

```python
rea = Reactome(root_data=root_data)

def open_tables(verbose:bool=True):
 
    _ = rea.open_reactome(verbose=verbose)
    _ = rea.open_reactome_abstract(verbose=verbose)
    _ = rea.open_reactome_gmt(verbose=verbose)

    return

def tool1(pathway: str, verbose:bool=True) -> Tuple[str, List]:

    df_gmt = rea.df_gmt

    row = df_gmt[df_gmt.pathway == pathway].iloc[0]

    genes = row.genes
    if isinstance(genes, str):
        genes = eval(genes)

    return row.pathway_id, genes
```

```python
verbose=True
open_tables(verbose=verbose)

len(rea.dfr), len(rea.pathway_list), rea.pathway_list[0]
```

```python
pathway="Innate Immune System"
verbose=True

pathway_id, genes = tool1(pathway, verbose=verbose)

print("")
print(pathway_id)
print(genes)
```
