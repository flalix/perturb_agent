### Environment


# in Dockerfile
ENV STORAGE_DIR=/opt/render/storage/data

```Python
from pathlib import Path
import os

def get_storage_dir():
    return Path(os.environ.get("STORAGE_DIR", "src/data"))

STORAGE_DIR = get_storage_dir()
```

### Run Local

```Bash
docker run -p 8501:10000 \
  -e STORAGE_DIR=/app/data \
  -v $(pwd)/data:/app/data \
  perturb-agent
```
