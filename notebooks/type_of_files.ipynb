### 🧬 GDC file type: **Aggregated Somatic Mutation**

In the Genomic Data Commons (GDC), an **Aggregated Somatic Mutation** file is a **processed mutation dataset** that combines and standardizes somatic variant calls across samples.

---

#### 🔎 What does it mean?

Instead of giving you raw variant calls per pipeline, this file:

* **Aggregates somatic mutations** detected in tumor samples
* Typically harmonizes results from variant callers (e.g., MuTect2)
* Produces a **clean, analysis-ready mutation table**
* Usually stored in **MAF format (Mutation Annotation Format)**


#### MuTect2 (GATK)

https://gatk.broadinstitute.org/hc/en-us/articles/360037593851-Mutect2


Call somatic SNVs and indels via local assembly of haplotypes

-----------

##### Category Short Variant Discovery

**Overview**

Call somatic short mutations via local assembly of haplotypes. Short mutations include single nucleotide (SNA) and insertion and deletion (indel) alterations. The caller uses a Bayesian somatic genotyping model that differs from the original MuTect by [Cibulskis et al., 2013](http://www.nature.com/nbt/journal/v31/n3/full/nbt.2514.html) and uses the assembly-based machinery of [HaplotypeCaller](https://gatk.zendesk.com/hc/en-us/articles/360037225632-HaplotypeCaller). Of note, Mutect2 v4.1.0.0 onwards enables joint analysis of multiple samples.

This tool is featured in the Somatic Short Mutation calling Best Practice Workflow. See [Tutorial#11136](https://software.broadinstitute.org/gatk/documentation/article?id=11136) for a step-by-step description of the workflow and [Article#11127](https://software.broadinstitute.org/gatk/documentation/article?id=11127) for an overview of what traditional somatic calling entails. For the latest pipeline scripts, see the [Mutect2 WDL scripts directory](https://github.com/broadinstitute/gatk/tree/master/scripts/mutect2_wdl). For pipelines with example data, see the [gatk-workflows repository](https://github.com/gatk-workflows/gatk4-somatic-snvs-indels). Although we present the tool for somatic calling, it may apply to other contexts, such as mitochondrial variant calling and detection of somatic mosaicism.

> “One consolidated table of all somatic mutations for a cohort or project”

#### 🧾 Typical format: MAF (Mutation Annotation Format)

These files follow a standardized schema widely used in cancer genomics (especially TCGA).

---

#### 📊 Main columns (most important ones)

Here are the key columns you’ll almost always see:

##### 🧬 Variant identification

* `Hugo_Symbol` → gene name (e.g., TP53)
* `Entrez_Gene_Id` → gene ID
* `Chromosome`
* `Start_Position`
* `End_Position`
* `Reference_Allele`
* `Tumor_Seq_Allele2` → mutated allele

---

##### 🧪 Mutation classification

* `Variant_Classification`

  * e.g., Missense_Mutation, Nonsense_Mutation, Silent
* `Variant_Type`

  * SNP, INS, DEL

---

##### 🧫 Sample information

* `Tumor_Sample_Barcode`
* `Matched_Norm_Sample_Barcode`

---

##### 🔬 Functional annotation

* `Protein_Change` (e.g., p.R175H)
* `Codon_Change`
* `Transcript_ID`

---

##### 📈 Quality / sequencing metrics

* `t_depth` → tumor read depth
* `t_ref_count` → reads supporting reference
* `t_alt_count` → reads supporting mutation

---

##### 🧪 Additional annotations (often present)

* `dbSNP_RS`
* `COSMIC_ID`
* `FILTER`
* `IMPACT` (LOW, MODERATE, HIGH)

---

#### 🧠 Why “Aggregated”?

Because GDC:

* Runs standardized pipelines across all samples
* Merges results into **one unified mutation dataset**
* Ensures **comparability across patients/projects**

---

#### 🧪 Practical use cases

This file is **central in cancer genomics pipelines**, e.g.:

* Mutation frequency per gene
* Driver gene discovery
* Tumor mutational burden (TMB)
* Pathway mutation enrichment
* Patient-specific mutation profiling (→ your perturbation framework 🔥)

---

#### ⚠️ Important notes

* It is **NOT raw data** → already processed and filtered
* It may **exclude low-confidence variants**
* Different GDC releases may slightly vary columns

---
