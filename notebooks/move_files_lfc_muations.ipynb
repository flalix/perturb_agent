```python
for i, row in df_psi.iterrows():
    psi_id = row['psi_id']
    gdc.set_primary_site(psi_id=psi_id)
    print("primary site", gdc.psi_id)

    files = os.listdir(gdc.root_disease)
    # print(";".join(files))
    if not files:
        continue

    #---------------- mutations -------------------------
    files2 = [x for x in files if x.startswith('mutations_') or x.startswith('vcf_')]

    if files2:
        print(files2)
        for fname in files2:
            shutil.move(gdc.root_disease / fname, gdc.root_mutations / fname)
    
    files2 = [x for x in files if x.startswith('samples_')]

    if files2:
        print(files2)
        for fname in files2:
            shutil.move(gdc.root_disease / fname, gdc.root_samples / fname) 

    files2 = [x for x in files if x.startswith('Gene_Expression_') or x.startswith('degs_') or x.startswith('lfc_')]
    
    if files2:
        print(files2)
        for fname in files2:
            shutil.move(gdc.root_disease / fname, gdc.root_lfc / fname)                
```
