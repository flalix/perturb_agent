## Fine Tuning

### Not alway the best procedure


https://www.linkedin.com/in/leo-wan/


The assumption that fine tuning always helps is deeply embedded in ML workflows. But a new study from Biogen across 18,540 antibodies from 33 drug programs shows that what you train on matters far more than how you train and fine tune.

They tested whether **AI language models that were trained on millions of protein sequences could predict which antibody candidates** would behave well in manufacturing and formulation. Three models and three assays measuring different biophysical risks.
 
#### Adapting antibody-specialist models to internal data improved predictions.
#### But adapting a general-purpose model made things worse. Fine-tuning and swapping models did not make any predictions better.

The more striking solution was simpler. When they added ~870 deliberately "bad" antibodies to the training set, containing sequences selected specifically for high off-target stickiness, the prediction accuracy improved across the board. This held even when testing on completely unrelated antibody targets the model had never seen.

#### This study suggests that a carefully composed dataset with well-represented "failure" data did more for generalization than swapping models and fine-tuning. 

The practical takeaway for antibody discovery teams: 
> Collect and curate your failures as carefully as your successes.   
> Teaching a model what bad looks like may be more valuable than giving it a better algorithm.
